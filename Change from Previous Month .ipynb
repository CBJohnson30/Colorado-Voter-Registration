Function to get all the counties percent change. 

```python
def percent_change_m(file): #"./Data/reg_data_tab.csv"
    perecent_change = pd.DataFrame(columns=['Active Dem', 'Active Rep', 'Active Uaf', 'Total Active',
       'Inactive Dem', 'Inactive Rep', 'Inactive Uaf', 'Total Inactive',
       'Grand Total', 'County','Label'])
    df = pd.read_csv(file)
    counties = df["County"].unique()
    df["Date"] = pd.to_datetime(df["Date"])
    df.sort_values("Date", inplace=True)
    df.set_index("Date", inplace=True)
    for county in counties:
        df_county = df[df["County"] == county]
        df_change = df_county.drop(["County",'Year', 'Month', 'Label'], axis=1)
        difference = df_change.pct_change()
        difference["County"] = df_county["County"]
        difference["Label"] = df_county["Label"]
        perecent_change = pd.concat([perecent_change, difference])
    return perecent_change
        
```

```python
change_from_prev_month = percent_change_m("./Data/reg_data_tab.csv")
```
